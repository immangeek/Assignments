**Importing the libraries**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

import seaborn as sns
import matplotlib.pyplot as plt
```

**Importing the dataset**

```python
data = pd.read_csv("Social_Network_Ads.csv")
data.head(5)
```

**Splitting the dataset into the Training set and Test set**

```python
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
X = data.drop(columns=['Purchased','User ID'],axis=1)
Y = data['Purchased']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

**Feature Scaling**

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

**Fitting Decision Tree Classification to the Training set**

```python
classifier = DecisionTreeClassifier()
```

```python
classifier.fit(x_train,y_train)
```

**Predicting the Test set results**

```python
ypred = classifier.predict(x_test)
```

**Making the Confusion Matrix**

```python
confusion_matrix(y_test,ypred)
```

**Visualising the Training set results**

```python
training_score = classifier.score(x_train,y_train)
training_score=training_score*100

print("Training data score is ---> ",training_score)
```

**Visualising the Test set results**

```python
#test_score = classifier.score(y_test,ypred)
test_score=accuracy_score(y_test,ypred)*100
#test_score=test_score*100

print("Test data score is --> ",test_score)
```
